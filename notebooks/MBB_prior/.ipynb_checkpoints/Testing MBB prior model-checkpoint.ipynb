What is the luminosity range and/or flux I get out with uniform priors on $N$ and $T$ and uniform on $\log N$ and $\log T$?

```python
model_code="""
data
{
    int nsrc;
}


transformed data {
  real k_B;
  real c;
  real h;
  k_B=1.3806488E-23;
  c=3.0E8;
  h=6.62606957E-34;

}




parameters {
  real<lower=0.0,upper=1E5> Nbb[nsrc];
  real<lower=10, upper=100.0> T[nsrc];
  real<lower=1.0,upper=2.0> Beta[nsrc];
  real<lower=1.4,upper=2.6> alpha[nsrc];
  real<lower=0.001,upper=5> z[nsrc];
}

transformed parameters {
  vector[nsrc] f_vec_psw;//vector of source fluxes
  vector[nsrc] f_vec_pmw;//vector of source fluxes
  vector[nsrc] f_vec_plw;//vector of source fluxes
  vector[nsrc] L_dust;


  for (i in 1:nsrc){

    f_vec_psw[i]=SPIRE_obs_no_filt(Nbb[i],alpha[i],Beta[i],T[i],z[i],c,h,k_B,250.0)*1E3;
    f_vec_pmw[i]=SPIRE_obs_no_filt(Nbb[i],alpha[i],Beta[i],T[i],z[i],c,h,k_B,350.0)*1E3;
    f_vec_plw[i]=SPIRE_obs_no_filt(Nbb[i],alpha[i],Beta[i],T[i],z[i],c,h,k_B,500.0)*1E3;
    L_dust[i]=log10(1.0E-23*LIR(Nbb[i],alpha[i],Beta[i],T[i],c,h,k_B)/3.826E33);
}

}
model{}
"""
```

```python
log_model_code="""
data
{
    int nsrc;
}


transformed data {
  real k_B;
  real c;
  real h;
  k_B=1.3806488E-23;
  c=3.0E8;
  h=6.62606957E-34;

}




parameters {
  real<lower=-4, upper=8> Nbb[nsrc];
  real<lower=-1, upper=5> T[nsrc];
  real<lower=1.0,upper=2.0> Beta[nsrc];
  real<lower=1.4,upper=2.6> alpha[nsrc];
  real<lower=0.001,upper=5> z[nsrc];
}

transformed parameters {
  vector[nsrc] f_vec_psw;//vector of source fluxes
  vector[nsrc] f_vec_pmw;//vector of source fluxes
  vector[nsrc] f_vec_plw;//vector of source fluxes
  vector[nsrc] L_dust;


  for (i in 1:nsrc){

    f_vec_psw[i]=SPIRE_obs_no_filt(exp(Nbb[i]),alpha[i],Beta[i],exp(T[i]),z[i],c,h,k_B,250.0)*1E3;
    f_vec_pmw[i]=SPIRE_obs_no_filt(exp(Nbb[i]),alpha[i],Beta[i],exp(T[i]),z[i],c,h,k_B,350.0)*1E3;
    f_vec_plw[i]=SPIRE_obs_no_filt(exp(Nbb[i]),alpha[i],Beta[i],exp(T[i]),z[i],c,h,k_B,500.0)*1E3;
    L_dust[i]=log10(1.0E-23*LIR(exp(Nbb[i]),alpha[i],Beta[i],exp(T[i]),c,h,k_B)/3.826E33);
}

}
model{}
"""
```

```python
cov_model_code="""
data
{
    int nsrc;
}


transformed data {
  real k_B;
  real c;
  real h;
  vector<lower=0.0>[3] mu;
  vector<lower=0.0>[3] S;

  k_B=1.3806488E-23;
  c=3.0E8;
  h=6.62606957E-34;
  mu[1]=0.0;
  mu[2]=3.7;
  mu[3]=1.6;
  S[1]=2.6;
  S[2]=0.8;
  S[3]=0.3;
  }





parameters {
  vector<lower=-4,upper=4>[3] alpha[nsrc];
  cholesky_factor_corr[3] L;
  real<lower=0.001,upper=5> z[nsrc];

}

transformed parameters {
 
  vector[3] beta[nsrc]; //N_bb,T,beta
  vector[nsrc] f_vec_green;//vector of source fluxes
  vector[nsrc] f_vec_red;//vector of source fluxes
  vector[nsrc] f_vec_psw;//vector of source fluxes
  vector[nsrc] f_vec_pmw;//vector of source fluxes
  vector[nsrc] f_vec_plw;//vector of source fluxes
  vector[nsrc] L_dust;
  for (i in 1:nsrc){
    beta[i] = exp(mu + S .* (L * alpha[i]));
    beta[i][3]=log(beta[i][3]);
    f_vec_green[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,100.0)*1E3;
    f_vec_red[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,160.0)*1E3;
    f_vec_psw[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,250.0)*1E3;
    f_vec_pmw[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,350.0)*1E3;
    f_vec_plw[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,500.0)*1E3;
    
    L_dust[i]=log10(1.0E-23*LIR(beta[i][1],2.0,beta[i][3],beta[i][2],c,h,k_B)/3.826E33);
}

}
model {
    for (i in 1:nsrc){
        alpha[i] ~ normal(0, 1);
        10.0 ~normal(L_dust[i],4);
}
L ~ lkj_corr_cholesky(1.0);
}
generated quantities {
  matrix[3,3] Omega;
  matrix[3,3] Sigma;
  Omega = multiply_lower_tri_self_transpose(L);
  Sigma = quad_form_diag(Omega, S); 
}
"""
```

```python
norm_model_code="""
data
{
    int nsrc;
}


transformed data {
  real k_B;
  real c;
  real h;
  k_B=1.3806488E-23;
  c=3.0E8;
  h=6.62606957E-34;

}




parameters {
  real<lower=8, upper=14> Nbb[nsrc];
  real<lower=-1, upper=5> T[nsrc];
  real<lower=1.0,upper=2.0> Beta[nsrc];
  real<lower=1.4,upper=2.6> alpha[nsrc];
  real<lower=0.001,upper=5> z[nsrc];
}

transformed parameters {
  vector[nsrc] f_vec_psw;//vector of source fluxes
  vector[nsrc] f_vec_pmw;//vector of source fluxes
  vector[nsrc] f_vec_plw;//vector of source fluxes
  vector[nsrc] L_dust;


  for (i in 1:nsrc){


    L_dust[i]=1.0E-23*LIR(1.0,alpha[i],Beta[i],exp(T[i]),c,h,k_B)/3.826E33;
    f_vec_psw[i]=pow(10.0,Nbb[i])*SPIRE_obs_no_filt(1.0,alpha[i],Beta[i],exp(T[i]),z[i],c,h,k_B,250.0)*1E3/L_dust[i];
    f_vec_pmw[i]=pow(10.0,Nbb[i])*SPIRE_obs_no_filt(1.0,alpha[i],Beta[i],exp(T[i]),z[i],c,h,k_B,350.0)*1E3/L_dust[i];
    f_vec_plw[i]=pow(10.0,Nbb[i])*SPIRE_obs_no_filt(1.0,alpha[i],Beta[i],exp(T[i]),z[i],c,h,k_B,500.0)*1E3/L_dust[i];
}

}
model{}
"""
```

## Fitting Prior model to required $L_dust$

```python
cov_model_code="""
data
{
    int nsrc;
}


transformed data {
  real k_B;
  real c;
  real h;


  k_B=1.3806488E-23;
  c=3.0E8;
  h=6.62606957E-34;
  }





parameters {
  vector<lower=-10,upper=10>[3] alpha[nsrc];
  cholesky_factor_corr[3] L;
  real<lower=0.001,upper=5> z[nsrc];
  real mu_logN;
  real<lower=0.0> S_logN;

}

transformed parameters {
 
  vector[3] beta[nsrc]; //N_bb,T,beta
  vector[nsrc] f_vec_green;//vector of source fluxes
  vector[nsrc] f_vec_red;//vector of source fluxes
  vector[nsrc] f_vec_psw;//vector of source fluxes
  vector[nsrc] f_vec_pmw;//vector of source fluxes
  vector[nsrc] f_vec_plw;//vector of source fluxes
  vector[nsrc] L_dust;
  vector<lower=0.0>[3] S;
  vector[3] mu;
  mu[1]=mu_logN;
  mu[2]=3.6;
  mu[3]=1.6;
  S[2]=0.25;
  S[1]=S_logN;
  S[3]=0.38;
  
  
  
  for (i in 1:nsrc){
    beta[i] = exp(mu + S .* (L * alpha[i]));
    beta[i][3]=log(beta[i][3]);
    f_vec_green[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,100.0)*1E3;
    f_vec_red[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,160.0)*1E3;
    f_vec_psw[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,250.0)*1E3;
    f_vec_pmw[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,350.0)*1E3;
    f_vec_plw[i]=SPIRE_obs_no_filt(beta[i][1],2.0,beta[i][3],beta[i][2],z[i],c,h,k_B,500.0)*1E3;
    
    L_dust[i]=log10(1.0E-23*LIR(beta[i][1],2.0,beta[i][3],beta[i][2],c,h,k_B)/3.826E33);
}

}
model {
    for (i in 1:nsrc){
        alpha[i] ~ normal(0, 1);
        10.0 ~normal(L_dust[i],2.5);
}
L ~ lkj_corr_cholesky(1.0);
//mu_logN~normal(0.0,0.1);
S_logN ~cauchy(0,5.0);


}
generated quantities {
  matrix[3,3] Omega;
  matrix[3,3] Sigma;
  Omega = multiply_lower_tri_self_transpose(L);
  Sigma = quad_form_diag(Omega, S); 
}
"""
cov_concat = ''.join([open('/Users/pdh21/HELP/XID_plus/stan_models/FIR_module.stan').read(),cov_model_code])
cov_sm = pystan.StanModel(model_code=cov_concat)
```

When fitting Arp220, it couldn't fit it.
After investigation, it appears as if there was a factor of $100^{2}$ out in the conversion from luminosity to flux (i.e. in the $D_{lum.}$ part. This caused too bigger Normalisation, which in turn caused too high luminosity. I am still a little concerned that Luminosity is out by 0.6 dec. 

```python
N_bb=0.54
beta=1.57
alpha=3.02
T=39.73
lam=np.arange(8,1000.0,1)

freq=3.0E14/lam
print np.log10(1.0E-23*np.trapz(Casey_SED(N_bb,beta,alpha,T,lam)[::-1],x=freq[::-1])/3.826E33)


Casey_to_flux(Casey_SED(N_bb,beta,alpha,T,350.0),350.0,0.018)
```

```python
def Casey_SED(N_bb,beta,alpha,T,lam):
    b_1=26.68
    b_2=6.246
    b_3=1.905E-4
    b_4=7.243E-5
    k_B=1.3806488E-23
    c=3.0E8
    h=6.62606957E-34
    S_bb=N_bb*np.power(c/(1E-6*lam),beta+3)/(np.exp(h*c/((1E-6*lam)*k_B*T))-1)
    L_a_T=np.power((np.power(b_1+b_2*alpha,-2.0)+(b_3+b_4*alpha)*T),-1.0)
    lam_c=3.0*L_a_T/4.0
    #----Casey 2012 function in table not working------
    #N_pl=N_bb*(1-np.exp(-1.0*(lam_0/lam_c)**beta))*np.power(lam_c,-3.0)/(np.exp(h*c/(lam_c*k_B*T))-1)
    #------Use this instead-------------
    A1 =(c/(lam_c*1E-6))**(beta+3.0)
    A2 = np.exp((h*c)/(1.0E-6*lam_c*k_B*T))-1.0
    B = (lam_c)**alpha
    N_pl = N_bb*A1/(A2*B)
    S_pl=N_pl*np.power(lam,alpha)*np.exp(-1.0*np.power(lam/lam_c,2.0))
    return (S_bb+S_pl)

def Casey_to_flux(flux,wave,z):
    from astropy.cosmology import Planck13
    import astropy.units as u
    D_l = -0.211 + 4431.95*z + 3329.47*np.power(z,2)-1047.52*np.power(z,3)
    return ((1.0+z)*flux/(4.0*np.pi*np.power(D_l*3.086E22,2)))#*np.power(wave*1.0E-6/(1.0+z),2)
```
